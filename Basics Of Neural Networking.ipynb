```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/drive/MyDrive/ML_Practice/Churn_Modelling.csv")
df
```

```python
x=df.iloc[:,3:-1].values
x
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x[:,1]=le.fit_transform(x[:,1])
x[:,2]=le.fit_transform(x[:,2])
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
ml=Sequential()
ml.add(Dense(units=64,activation='relu',input_dim=10))
ml.add(Dense(units=32,activation='relu'))
ml.add(Dense(units=16,activation='relu'))
ml.add(Dense(units=1,activation='sigmoid'))
ml.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
ml.summary()

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
ml.fit(x_train,y_train,epochs=100,batch_size=32)
```
